# **Exploratory Data Analysis (EDA)** 
Performing Exploratory Data Analysis (EDA) to identify forward strikers born after the year 2000 in the **French Ligue 1**.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#Player Standard Stats
PSS = 'Player_Standard_Stats2.csv'
#Player Standard Stats
PP = 'Player_Passing.csv'
#Squad Standard Stats
SSS = 'Squad_Standard_Stats.csv'
#Player_Goal_and_Shot_Creation
Goals_shoots = 'Shoots.csv'
```

```python
data1 = pd.read_csv(PSS)
#SSS = pd.read_csv(SSS,skiprows=1)
passes = pd.read_csv(PP,skiprows=1)
goals = pd.read_csv(Goals_shoots,skiprows=0)
```

Drop unneeded columns like nation, age (born year already exist), number (RK)

Drop duplicates as well

```python
cols1 = data1.columns.drop(["Nation","Rk"])
```

```python
PSS = data1[cols1].drop_duplicates()
goals1 = goals.columns.drop(["Nation","Rk"])
goals = goals[goals1].drop_duplicates()
```

```python
passes_1 = passes.columns.drop(["Nation","Rk"])
passes= passes[passes_1].drop_duplicates()
```

```python
passes
```

```python
Year = 2000
PSS = PSS[PSS['Born']>Year]
PSS
```

```python
filtered_players = PSS[~PSS['Pos'].isin(['GK', 'DF','MF,DF','MF'])]
filtered_players
```

## **2. Exploratory Data Analysis**


```python
# Group players by club and count the number of players under 21 in each club
club_counts = filtered_players.groupby('Squad')['Player'].count()

# Sort clubs by the number of under 21 players in descending order
club_counts_sorted = club_counts.sort_values(ascending=False)
```

```python
plt.figure(figsize=(10, 4))
club_counts_sorted.plot(kind='bar', color='skyblue')
plt.title('Clubs with the Most Under 21 Players')
plt.xlabel('Club')
plt.ylabel('Number of Under 21 Players')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
sorted_df = filtered_players.sort_values(by='MP', ascending=False)
sorted_df
```
